# Pandas Reshape Tutorial

Using: [link](https://www.w3resource.com/pandas/reshaping.php)

```python
import numpy as np
import pandas as pd
```

```python
tuples = list(zip(*[['green', 'green', 'blue', 'blue', 'red', 'red'],
                    ['one'  , 'two'  , 'one' , 'two' , 'one', 'two']]))


index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
df = pd.DataFrame(np.random.randn(6, 2), index=index, columns=['M', 'N'])
df_orig  = df.copy()

df
```

```python
# stack
df_stacked = df.stack()
df_stacked
```

```python
df_stacked.unstack([0])
```

### Pivot Tables

```python
# pivot tables

df = pd.DataFrame({'M': ['one', 'one', 'two', 'three'] * 2,
                   'N': ['A', 'B'] * 4,
                   'O': ['blue', 'blue', 'red', 'red'] * 2,
                   'P': np.random.randn(8),
                   'Q': np.random.randn(8)})

df
```

```python
pd.pivot_table(df,index=['O','N'],columns=['M'])
```
