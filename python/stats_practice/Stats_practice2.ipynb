##### Importing Libraries

```python
# general
import sys
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()

# mpl
import matplotlib as mpl
import matplotlib.pyplot as plt

# plotly
import plotly
import plotly.express as px
import plotly.io as pio
import plotly.figure_factory as ff
import plotly.graph_objects as go
pio.renderers.default = 'iframe'

# stats
from scipy import stats
from lmfit.models import GaussianModel

# my helper functions
sys.path.insert(0, '../helpers/')
import pd_helpers
import plotly_helpers
```

```python
# drawing a normal distribution

# config
mu = 0
sigma = 1

# sample
x = np.linspace(mu-(3*sigma),mu+(3*sigma),num=200)
y = stats.norm.pdf(x,loc=mu,scale=sigma)

# draw
fig = px.line(x=x,y=y)
fig.show()

```

```python
# random gaussian distribution

# config
mu = 0 
sigma = 1

# sample
y = np.random.normal(mu,sigma,int(1e4))
```

```python
# plot histo
fig2 = px.histogram(y,marginal='box',height=400,width=600,nbins=80)
fig2.show()
```

```python
# plot with gaussian line

label = 'Gaussian, mu: {}, sigma: {}'.format(mu,sigma)

fig3 = ff.create_distplot([y],[label],curve_type='normal',bin_size=.1)
fig3.show()
```

```python
# get gaussian fit 
model = GaussianModel()

# initialise
params = model.make_params(center=mu,amplitude=200,sigma=1)  

# make histogram
h, x_edges = np.histogram(y,bins=80)
x_centers = (x_edges[1:] + x_edges[:-1])/2

# fit 
fit = model.fit(h,params,x=x_centers)

# draw
fig5 = px.line(x=x_centers,y=fit.best_fit)
fig5.show()
```

```python
# get fit results 
print(fit.fit_report())
```
