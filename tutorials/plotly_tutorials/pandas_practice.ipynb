#### 1. Date difference

```python
# get data
museum_path = '../data/datavis/museum_visitors.csv'
museum_data = pd.read_csv(museum_path)
```

```python
museum_data.tail()
```

```python
museum_data.dtypes
```

```python
# parse date
museum_data['Date'] = pd.to_datetime(museum_data.Date,format="%Y-%m-%d")
museum_data.dtypes
```

```python
today = datetime.today()
today = pd.to_datetime('2019-01-01',format='%Y-%m-%d')

museum_data['days_elapsed'] = (today - museum_data.Date).dt.days

museum_data.sort_values('days_elapsed', inplace=True, ascending=False)
museum_data.head()
```

```python
aa_atendance = museum_data['Avila Adobe'].groupby(museum_data.Date.dt.year).mean()
```
