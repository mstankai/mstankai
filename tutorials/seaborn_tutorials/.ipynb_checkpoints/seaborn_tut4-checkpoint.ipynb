```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
# set resolution
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
```

```python
# get data
insurance_path = "./data/datavis/insurance.csv"
insurance = pd.read_csv(insurance_path)
```

```python
# data overview
display(insurance.head())
print("Shape: ",insurance.shape)
```

```python
# plotting BMI agains insurance charges

# defining plot
fig, ax = plt.subplots(figsize=(4,4))

# making scatter plot
sns.scatterplot(x=insurance.bmi
                ,y=insurance.charges
                ,ax=ax)
# cosmetics
ax.set_title("BMI vs Insurance Charges")
ax.set_xlabel("BMI")
ax.set_ylabel("Insurance Charge, USD")
```

```python
# ==================================================
# funtion drawing fitresult in graph 
# -- authored by me (Migle Stankaityte)

def draw_LR_fitresult(x,y,ax
                      ,debug=False
                      ,fsize=12
                      ,tag=""
                      ,loc=(0.05,0.95)) :

    # fit details

    
    fit_res = stats.linregress(x,y)
    
    # important values
    # Y = aX + c
    a, c, R, p, a_err = stats.linregress(x,y)
    c_err = fit_res.intercept_stderr
    
    # R suared 
    R_sq = R**2
    
    # write strings
    fitres_str = {
        "a"      : "{:5.2f} +- {:5.2f}".format(a,a_err),
        "c"      : "{:5.2f} +- {:5.2f}".format(c,c_err),
        "R^2"    : "{:.2f}%".format(R_sq*100),
        "p val." : "{:.0e}".format(p)
    }
    
    if debug:
        for key, val in fitres_str.items() : print(key+"\t"+val)
    
    # define text as single string
    fit_info_str = "Fit Results: {:s}\n".format(tag)
    
    for key, val in fitres_str.items() : 
        fit_info_str += "\n" + key + ":   " + val
        

    # box properties - cosmetic
    box_props = { 'boxstyle' : 'square', 
                  'facecolor' : 'white', 
                  'alpha' : 0.5
                }

    
    # draw text box
    ax.text(loc[0], loc[1] 
            ,fit_info_str 
            ,transform=ax.transAxes 
            ,fontsize=fsize
            ,verticalalignment='top' 
            ,bbox=box_props)
    
    
# ==================================================
```

```python
# plotting BMI agains insurance charges
# -->  with regression

# defining plot
fig, ax = plt.subplots(figsize=(8,8))

# making scatter plot 
sns.regplot(x=insurance.bmi
            ,y=insurance.charges
            ,ax=ax)

# cosmetics
ax.set_title("BMI vs Insurance Charges")
ax.set_xlabel("BMI")
ax.set_ylabel("Insurance Charge, USD")

# draw fit result box
draw_LR_fitresult(x=insurance.bmi
                 ,y=insurance.charges
                 ,ax=ax)

```

```python
# color coding same plot by smoking status
fig, ax = plt.subplots(figsize=(6,6))

# making plot
sns.scatterplot(x=insurance.bmi
                ,y=insurance.charges
                ,hue=insurance.smoker
                ,ax=ax)


# cosmetics
ax.set_title("BMI vs Insurance Charges")
ax.set_xlabel("BMI")
ax.set_ylabel("Insurance Charge, USD")


```

```python
smokers = insurance.loc[insurance.smoker=='yes']
non_smokers = insurance.loc[insurance.smoker=='no']
```

```python
# color coding same plot by smoking status
# --> with regresion

# making plot
lm = sns.lmplot(data=insurance
           ,x='bmi'
           ,y='charges'
           ,hue='smoker'
           ,height=6
           ,aspect=2
           ,legend=False) 

# plotting legend inside figure
plt.legend(title='Smoker', 
           loc='upper left',
           bbox_to_anchor=(0.25,0.971))


# cosmetics
plt.title("BMI vs Insurance Charges")
plt.xlabel("BMI")
plt.ylabel("Insurance Charge, USD")

# get axes
ax = lm.axes[0,0]

# draw fit result box -- smokers
draw_LR_fitresult(x=smokers.bmi
                 ,y=smokers.charges
                 ,tag="Smokers"
                 ,ax=ax)



# draw fit result box -- non smokers
draw_LR_fitresult(x=non_smokers.bmi
                 ,y=non_smokers.charges
                 ,tag="Non Smokers "
                 ,loc=(0.05,0.72)
                 ,ax=ax)



```

# Trying Plotly

Now I will try to make the same plot using plotly

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
# make plot
fig = px.scatter(insurance
                ,x='bmi'
                ,y='charges'
                ,color='smoker'
                ,trendline='ols') 

# draw plot
fig.show()
```

```python
# get fit results
results = px.get_trendline_results(fig)

# print fit results
print('SMOKER fit results\n')
display(results.query("smoker == 'yes'").px_fit_results.iloc[0].summary())
print('------------')
print('\nNON SMOKER fit results\n')
results.query("smoker == 'no'").px_fit_results.iloc[0].summary()




```
