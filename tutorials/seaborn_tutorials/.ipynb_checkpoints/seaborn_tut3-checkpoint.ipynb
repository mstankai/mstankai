```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set higher resolution
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 300
```

```python
# get data
flight_filepath = "./data/datavis/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
# view data
display(flight_data.head())
flight_data.describe()
```

```python
# bar chart for NK flight delays
plt.figure(figsize=(6,4))

# plot
sns.barplot(x=flight_data.index,
            y=flight_data.NK)

# cosmetics
plt.title("Average monthly delay for NK airlines")
plt.xlabel("Month")
plt.ylabel("Delay, min")
```

```python
# make a "heatmap" (2d histo) showing all data

# create figure
plt.figure(figsize=(10,5))

# make plot
sns.heatmap(data=flight_data 
            ,annot=True
            ,annot_kws={'size':8}
            ,cmap="viridis")

# cosmetics
plt.title("Average arival delay per Month")
plt.xlabel("Airline")
plt.ylabel("Month")
```
